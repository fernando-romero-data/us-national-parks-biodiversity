# 05 – Project Report: Biodiversity in U.S. National Parks

This notebook summarizes the key insights from the biodiversity analysis across four U.S. National Parks.  
It answers four core questions related to conservation status, species vulnerability, statistical significance, and park-level trends.

## Goal
Analyze species and observation data to understand biodiversity patterns, identify which species groups are most protected, and determine whether differences across categories are statistically meaningful.

## Findings

### 1. Conservation Status Distribution
- **5,633 species** → No Intervention  
- **191 species** → Under Protection  

Only a small fraction of species are part of conservation programs, showing a highly imbalanced protection distribution.


### 2. Which Species Are Most Protected?
- **Mammals** and **Birds** have the **highest protection rates**.
- These species categories are more likely to be endangered or prioritized for conservation action.


### 3. Are Differences Between Categories Significant?

Using chi-square independence tests:

- **Mammals vs Birds** → *Not significant*  
- **Mammals vs Reptiles** → *Significant difference*  

This means Mammals and Reptiles have **meaningfully different conservation rates**, while Mammals and Birds do not.


### 4. Which Animals Stand Out?
- **Bats** are the most frequently protected group across the dataset.
- **Yellowstone National Park** contains the **largest protected bat population**.

These trends highlight bats as a notable conservation priority in the parks examined.

## Methods
- Data cleaning and grouping using `pandas`
- Conservation flags: `is_protected`, `is_bat`
- Aggregation by species category and park
- Chi-square significance testing (`scipy.stats`)
- Visualization using `seaborn`


## Conclusion
Most species across the parks are not under conservation, but certain groups — especially **Mammals, Birds, and Bats** — receive higher levels of protection.

Bat populations are particularly strong in **Yellowstone**, making it a key location for conservation focus.

This project offers a clear, data-driven view of conservation priorities within U.S. National Parks.
